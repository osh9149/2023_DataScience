```python
import csv
f=open('2020_temp.csv','r',encoding='cp949')
lines=csv.reader(f)
header=next(lines)
max_temp=0
min_temp=100
max_date=''
min_date=''

for line in lines:
    if max_temp<float(line[4]):  #최고기온
        max_temp=float(line[4])
        max_date=line[0]

    if min_temp>float(line[3]):  #최저기온
        min_temp=float(line[3])
        min_date=line[0]

print('최고기온 날짜와 온도 : ', max_date, max_temp)
print('최저기온 날짜와 온도 : ', min_date, min_temp)
f.close()
```

```python
import csv
import matplotlib.pyplot as plt
from matplotlib import rc
plt.rc('font', family='NanumBarunGothic') 
f=open('seoul.csv', 'r', encoding='cp949')
lines=csv.reader(f)
next(lines)
max_result=[]
min_result=[]
avg_result=[]

for line in lines:
    if line[4] != '': #최고기온이 빈칸이 아닐때
        max_result.append(float(line[4]))

    if line[3] != '': #최저기온이 빈칸이 아닐때
        min_result.append(float(line[3]))

    if line[2] != '': #평균기온이 빈칸이 아닐때
        avg_result.append(float(line[2]))

#print(max_result)
plt.figure(figsize=(10, 3))
plt.plot(max_result, 'r')
plt.plot(min_result, 'b')
plt.plot(avg_result, 'g')
plt.title('1907년~2021년 기온데이터 분석')
plt.show()
f.close()



```
