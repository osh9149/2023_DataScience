```python
import csv
import numpy as np

f=open('2021_age1.csv','r',encoding='CP949')
lines=csv.reader(f)
next(lines)
lines=list(lines)
mn=1
result_name=''
result=0
name=input('궁금한 지역이름을 입력하세요 : ')

for line in lines:
    if name in line[0]:
        home=np.array(line[3:], dtype=int)/int(line[2])
# print(home)
for line in lines:
    away=np.array(line[3:], dtype=int)/int(line[2])
# print(home-away)
    s=np.sum((home-away)**2)
    if s<mn and name not in line[0]:
        mn=s
        result_name=line[0]
        result=away
print(result_name)
addr1=result_name.split('(')
print(addr1[0])
addr2=addr1[0].split(' ')
same=addr2[-1]
f.close()

```

```python

plt.figure(figsize=(10,5), dpi=300)
plt.title(name+'지역과 가장 비슷한 인구구조를 가진 지역')
plt.plot(home, label=name)
plt.plot(result,label=same)
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic') 
f=open('2021_gender1.csv','r',encoding='CP949')
lines=csv.reader(f)
male=[]
female=[]
name=input('알고싶은 동이름을 입력하세요 : ')
for line in lines:
    if name in line[0]:
        for i in line[3:104]:
            if i != '':
                male.append(-int(i))
        for i in line[106:207]:
            if i != '':
                female.append(int(i))

plt.title(name+'의 인구구조 분석')
plt.barh(range(101), male, label='남성')
plt.barh(range(101), female, label='여성')
plt.legend()
plt.show()
f.close()
```
