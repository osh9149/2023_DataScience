#문제4. 서울의 와이파이 현황 파일(seoul_wifi.xlsx)을 판다스 라이브러리를 이용해 읽는다.(5점)

```python
#문제4번(5점)

df = pd.read_excel('seoul_wifi.xlsx', thousands = ',')
df
```

#문제5. 자치구별로 와이파이 설치 개수를 카운트하여 내림차순으로 새로운 데이터프레임(df1)으로 만든다.(5점)


```python
df1=df.groupby('자치구').count()
df1=df1.sort_values(by=['와이파이명'], ascending=[False])
df1
```

#문제6. df1에서 와이파이가 가장 많이 설치된 곳과 가장 적게 설치 된 곳을 찾아 아래와 같이 출력하시오.(5점)

```python

df1['와이파이명'].idxmax()
```

```python
df1['와이파이명'].idxmin()
```

```python
print('서울시에서 와이파이가 가장 많이 설치된 곳은 %s이며 가장 적게 설치된 곳은 %s입니다.'%(df1['와이파이명'].idxmax(),df1['와이파이명'].idxmin()))
```

#문제7. 위의 데이터프레임(df1)을 그래프로 그림처럼 시각화한다.(5점)
20230531_075007.png

```python
plt.rc('font', family='NanumBarunGothic') 
plt.figure(figsize=(15, 5))
plt.bar(df1.index, df1['와이파이명'], color='red')

plt.title('서울시 자치구별 와이파이 설치 현황', fontsize=20)
plt.xlabel('자치구', fontsize=14)
plt.ylabel('개수',rotation='vertical',fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

#문제8. df의 데이터 중 설치유형이 '버스정류소' 또는 '한강'인 데이터만 추출하여 새로운 데이터프레임(df2)으로 만든다

```python
#연산자 | - 또는 or
#연사자 & - 그리고 and

df2=df[df['설치유형'].str.contains('버스정류소') | df['설치유형'].str.contains('한강')]

df2
```

#문제8. df2의 데이터 중 원하는 '자치구' 이름을 입력받아 해당하는 자치구의 데이터만 새로운 데이터프레임(df3)로 만든다.(5점)
ex)성동구 입력하여 확인

```python
name=input('자치구 이름을 입력하세요 : ')
df3=df2[df2['자치구'].str.contains(name, na=False)]
df3
```

#문제9. df3 의 버스정류소 와이파이 위치를 지도에 마커로 표시(zoom_start=13)하고 팝업에 이름(와이파이명)이 보이도록 한다.(5점)

문제11. 데이터프레임의 위도, 경도 데이터를 이용하여 지도로 시각화한다.(10점)20230523_121650.png

```python
#문제10번(10점)

#새로 만든 데이터프레임(df3)에서 첫번째 데이터의 위도, 경도로 맵의 시작위치를 지정함
s_lat = df3.loc[df3.index[0], 'X좌표'] 
s_lng = df3.loc[df3.index[0], 'Y좌표']

map = folium.Map(
    location=[s_lat,s_lng ], tiles ='OpenStreetMap' , zoom_start=13)

for idx in df3.index:

    lat = df3.loc[idx, 'X좌표']
    lng = df3.loc[idx, 'Y좌표']
    name = df3.loc[idx, '와이파이명']
    
    folium.Marker(
        location=[lat, lng], 
				popup=name
    ).add_to(map)

map
```
