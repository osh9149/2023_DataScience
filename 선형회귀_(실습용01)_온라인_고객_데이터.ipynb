### <5> sklearn(인공지능 라이브러리)으로 가장 훌륭한 예측선 구하기

선형회귀는 sklearn.linear_model 안의 LinearRegression



```
X = data3["x값"]
y = data3["실제 y값"]
plt.plot(X, y, 'o')
plt.show()
```

```
from sklearn.linear_model import LinearRegression

line_fitter = LinearRegression()

line_fitter.fit(X.values.reshape(-1,1), y)
```

````python

# 선형회귀 연산이 가능한 모델 만들어주기

# 모델에 데이터를 넣고 학습시키기

# 모델이름.fit(x축데이터, y축데이터)
````

가장 정확한 예측선이 model에 담겨있음


*   기울기 : model.coef_
*   y절편 : model.intercept_



```
print(line_fitter.coef_) #기울기, 가중치, W
print(line_fitter.intercept_) #절편, 바이어스, b
```

````python
# 가장 정확한 예측선의 기울기

# 가장 정확한 예측선의 y절편
````

###기울기와 절편으로 만든 예측 함수에 따라 그래프를 그려보자
